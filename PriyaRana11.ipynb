```python
#fabonaccai by def function
def fabonaccai(n):
  first_number=0
  second_number=1
  for i in range(3,n+1):
   third_number=first_number+second_number
   first_number=second_number
   second_number=third_number
  return third_number
fabonaccai(10)
```

```python
#hcf by def function
def hcf(a,b):
  if a>b:
    smaller=b
  else:
    smaller=b
  for i in range(1,smaller+1):
    if a%i==0 and b%i==0:
      hcf=i
  return hcf
hcf(5,20)

```

```python
#Lcm by def function # while loop
def lcm (a,b):
  if a>b:
    largest=a
  else:
    largest=b
    while True:
     if largest%a==0 and largest%b==0:
      lcm=largest
      break
      largest+=1
    return lcm
lcm(5,20)



```

```python
#lcm by def function # for loop
def lcm (a,b):
  if a>b:
    largest=a
  else:
    largest=b
  for i in range(largest,a*b+1):
    if i % a==0 and i % b==0:
      lcm=i
      break
  print(lcm)
lcm(20,40)
lcm(30,60)
```
