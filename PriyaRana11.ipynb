```python
#Average
a=(5,20,25,30)
total_average=0
for i in range(len(a)):
  total_average+=a[i]
total_average=total_average/len(a)
print(total_average)



```

```python
b=(26,5,6,2,3) # odd number
odd_numbers = []
for num in b:
  if num % 2 != 0:
    odd_numbers.append(num)
print(odd_numbers)
```

```python
#product
c=(3,2,4,1,6)
total_product=1
for i in range(len(c)):
  total_product*=c[i]
print(total_product)
```
